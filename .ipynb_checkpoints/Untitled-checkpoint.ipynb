```python
#Import data
df = pd.read_csv (r'data/salaries.csv')
print (df)
```

```python
salaries = df[['salary']]
salaries.rolling(100000).mean().plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('employee', fontsize=20);
```

###  We can observe that managers (110,000-112,000) are at the bottom of salaries...

```python
# calculate summary statistics
salaries['salary'].copy()

data_mean=np.mean(salaries).copy() 
print('--------------------------------------')
print('MEAN AND STANDARD DEVIATION')
print('--------------------------------------')
print('MEAN OF AVERAGE SALARIES')
print(data_mean)

data_std=np.std(salaries)
print('STANDARD DEVIATION OF SALARIES')
print(data_std)
```

###  It means, managers earn average salary. (not logical at all)

```python
#Scatter plot
plt.figure(figsize=(20,15))
plt.title('Distribution?', fontsize=30)
plt.scatter(df['salary'],df['from_date'])
ax=plt.axes()
plt.xticks(fontsize= 30, rotation=25)
ax.xaxis.set_major_locator(plt.MaxNLocator(10))
plt.yticks(fontsize= 30)
ax.yaxis.set_major_locator(plt.MaxNLocator(10))
plt.show()
```

### At the same time we can see in this distribution, salaries has the same pattern during 13th years...
